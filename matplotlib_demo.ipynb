# MATPLOTLIB DEMONSTRATION

This Notebook file will demonstrate the basic operations which are used for the Visualizations in general. This is not complete by any means but will just cover the brief overview of the Matplot lib for beginers

Imports used for the Matlplot library usage 

```python
import matplotlib.pyplot as plt 
from matplotlib import style 
import random
%matplotlib inline 
```

# SAMPLE GRAPH WITH RANDOM DATA 

```python
# This is just to generate a sample data for a given range 
sample_population = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('ggplot')

# create the plot based on the above data
plt.plot(year_of_collectiontime_hrs,sample_population)

#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

# This is finally to show the plot as a picture 
plt.show()
```

# MODIFIED THE STYLE OF BACKGROUND AND GRAPH LINE 

```python
# This is just to generate a sample data for a given range 
sample_population = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('dark_background')

# create the plot based on the above data with different style 
plt.plot(year_of_collectiontime_hrs,sample_population,color='y',linestyle='--',linewidth=2.5)

#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

# This is finally to show the plot as a picture 
plt.show()
```

# DIFFERENT STYLE AND LINE AGAIN

```python
# This is just to generate a sample data for a given range 
sample_population = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('grayscale')

# line width is modified here as compared to above variants 
plt.plot(year_of_collectiontime_hrs,sample_population,color='b',label='Web traffic',linewidth=1.0)

#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

# This is finally to show the plot as a picture 
plt.show()
```

```python
# This is just to generate a sample data for a given range 
sample_population = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('seaborn-whitegrid')

plt.plot(year_of_collectiontime_hrs,sample_population,alpha=0.4)

#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')


plt.show()
```

# MULTIPLE GRAPHS IN SAME WINDOW

```python
# This is just to generate a sample data for a given range, but having more that one data collection
sample_population_country_1 = [random.randint(10, 5000) for iter in range(14)]
sample_population_country_2 = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('seaborn-poster')

#fitting both the data sets in to the same plot
plt.plot(year_of_collectiontime_hrs,sample_population_country_1,color='b',label='Monday',linewidth=1.0)
plt.plot(year_of_collectiontime_hrs,sample_population_country_2,color='r',label='Tuesday',linewidth=1.5)


#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

#showing that plots finally
plt.show()
```
