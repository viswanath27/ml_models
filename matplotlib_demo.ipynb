# MATPLOTLIB DEMONSTRATION

# SAMPLE GRAPH WITH RANDOM DATA 

```python
import matplotlib.pyplot as plt 
from matplotlib import style 
import random
%matplotlib inline 

# This is just to generate a sample data for a given range 
sample_population = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('ggplot')

# create the plot based on the above data
plt.plot(year_of_collectiontime_hrs,sample_population)

#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

# This is finally to show the plot as a picture 
plt.show()
```

# MULTIPLE GRAPHS IN SAME WINDOW

```python
import matplotlib.pyplot as plt 
from matplotlib import style 
import random
%matplotlib inline 

# This is just to generate a sample data for a given range, but having more that one data collection
sample_population_country_1 = [random.randint(10, 5000) for iter in range(14)]
sample_population_country_2 = [random.randint(10, 5000) for iter in range(14)]

# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

# Assign a style to the plot
style.use('seaborn-poster')

#fitting both the data sets in to the same plot
plt.plot(year_of_collectiontime_hrs,sample_population_country_1,color='b',label='country_1',linewidth=1.0)
plt.plot(year_of_collectiontime_hrs,sample_population_country_2,color='r',label='country_2',linewidth=1.5)


#Below are just to add up more info to the graph
plt.title('Population view of the Data')
plt.xlabel('Years')
plt.ylabel('sample quantity')

#showing that plots finally
plt.show()
```


# DRAWING SUBPLOTS FOR DATA 

```python
import matplotlib.pyplot as plt 
from matplotlib import style 
import random
%matplotlib inline 

# This is just to generate a sample data for a given range, but having more that one data collection
Parameter_1 = [random.randint(10, 5000) for iter in range(14)]
Parameter_2 = [random.randint(10, 5000) for iter in range(14)]
Parameter_3 = [random.randint(10, 5000) for iter in range(14)]
Parameter_4 = [random.randint(10, 5000) for iter in range(14)]
# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

plt.figure(figsize=(20,8))
plt.subplots_adjust(hspace=.5)
plt.subplot(2,2,1)
plt.title('Parameter_1')
plt.plot(year_of_collectiontime_hrs,Parameter_1)
plt.subplot(2,2,2)
plt.title('Parameter_2')
plt.plot(year_of_collectiontime_hrs,Parameter_2)
plt.subplot(2,2,3)
plt.title('Parameter_3')
plt.plot(year_of_collectiontime_hrs,Parameter_3)
plt.subplot(2,2,4)
plt.title('Parameter_4')
plt.plot(year_of_collectiontime_hrs,Parameter_4)
```

# OVERWRITING THE PLOTS ON EACH OF THEM FOR A SPECIFIC VIZUALIZATION

```python
import matplotlib.pyplot as plt 
from matplotlib import style 
import random
%matplotlib inline 

# This is just to generate a sample data for a given range, but having more that one data collection
Parameter_1 = [random.randint(10, 5000) for iter in range(14)]
Parameter_2 = [random.randint(10, 5000) for iter in range(14)]
Parameter_3 = [random.randint(10, 5000) for iter in range(14)]
Parameter_4 = [random.randint(10, 5000) for iter in range(14)]
# This is the year in which sample is collected
year_of_collectiontime_hrs=[2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

plt.figure(figsize=(20,8))
plt.subplots_adjust(hspace=.5)
plt.subplot(2,2,1)
plt.title('All Parameters together')
plt.plot(year_of_collectiontime_hrs,Parameter_1,color='b',linewidth=1.0)
plt.plot(year_of_collectiontime_hrs,Parameter_2,color='r',linewidth=1.0)
plt.plot(year_of_collectiontime_hrs,Parameter_3,color='g',linewidth=1.0)
plt.plot(year_of_collectiontime_hrs,Parameter_4,color='y',linewidth=1.0)
```
