```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# 2 - LOAD BOSTON DATA

# 3 - CONVERT TO PANDAS DATAFRAME

### 3.2 - DATA OVERVIEW
<p>Attribute Information:</p>
<ol>
    <li><b>CRIM</b>      &emsp;per capita crime rate by town</li>
    <li><b>ZN</b>        &emsp;proportion of residential land zoned for lots over 25,000 sq.ft.</li>
    <li><b>INDUS</b>     &emsp;proportion of non-retail business acres per town</li>
    <li><b>CHAS</b>      &emsp;Charles River dummy variable (= 1 if tract bounds river; 0 otherwise)</li>
    <li><b>NOX</b>       &emsp;nitric oxides concentration (parts per 10 million)</li>
    <li><b>RM</b>        &emsp;average number of rooms per dwelling</li>
    <li><b>AGE</b>       &emsp;proportion of owner-occupied units built prior to 1940</li>
    <li><b>DIS</b>       &emsp;weighted distances to five Boston employment centres</li>
    <li><b>RAD</b>       &emsp;index of accessibility to radial highways</li>
    <li><b>TAX</b>       &emsp;full-value property-tax rate per &#36;10,000</li>
    <li><b>PTRATIO</b>   &emsp;pupil-teacher ratio by town</li>
    <li><b>B</b>         &emsp;1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town</li>
    <li><b>LSTAT</b>     &emsp;  % lower status of the population</li>
    <li><b>MEDV</b>      &emsp;Median value of owner-occupied homes in &#36;1000's</li>
</ol>

```python
pd_boston_data = pd.DataFrame(boston_data.data,columns=['CRIM','ZN','INDUS','CHAS','NOX',
                                                        'RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT'])
```

```python
pd_boston_data
```

As all the data is in numbers and we do not have any text data which needs encoding from text to numeric we can directly apply regression. But before that we have to split the data as test and train.

# BASIC DATA SUMMARY

```python
pd_boston_data.describe()
```

# CHECK FOR NAN VLAUES IF ANY

```python
pd_boston_data.isnull().values.any()
```

# SEPARATE THE FEATURES AND TARGET

```python
features_x = pd_boston_data
features_x
```

```python
target = pd.DataFrame(boston_data.target,columns=['MEDV'])
target
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(features_x, target, test_size=0.33, random_state=42)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
train_y.shape
```

```python
test_y.shape
```

```python
test_y
```

# APPLY REGGRESSION

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_x, train_y)
```

```python
reg.score(train_x, train_y)
```

```python
y_pred = reg.predict(test_x)
```

```python
pd_results = pd.DataFrame(y_pred,columns=['y_pred'])
```

```python
pd_results
```
